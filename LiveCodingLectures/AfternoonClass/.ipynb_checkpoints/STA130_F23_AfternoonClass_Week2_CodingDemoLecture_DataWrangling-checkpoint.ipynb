```python
ab_non_nan = ab.dropna()#.isnull()
```

```python
ab_non_nan
```

```python
ab_non_nan['Pub year'] = ab_non_nan['Pub year'].astype(int)
```

```python
ab_non_nan.dtypes
```

```python
ab_non_nan['is_paper'] = (ab_non_nan.Hard_or_Paper=='P')
ab_non_nan.dtypes
```

```python
ab_non_nan.Hard_or_Paper = ab_non_nan['Hard_or_Paper'].astype("category")
ab_non_nan.dtypes
```

```python
ab_non_nan.groupby("Hard_or_Paper")
```

```python
ab_non_nan.groupby("Hard_or_Paper").mean(numeric_only=True)
```

```python
ab_non_nan.is_paper.mean()
```

```python
ab_non_nan.groupby("Hard_or_Paper").sum()
```

```python
ab_non_nan[['NumPages','Thick','Weight_oz','Hard_or_Paper']].groupby('Hard_or_Paper').max()
```

```python
ab_non_nan.groupby('Hard_or_Paper').size()
```

```python
ab_non_nan.groupby('Hard_or_Paper').count()
```
