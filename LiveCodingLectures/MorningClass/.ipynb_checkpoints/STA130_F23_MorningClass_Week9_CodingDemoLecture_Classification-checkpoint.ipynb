````python
import pandas as pd
ab = pd.read_csv("../../amazonbooks.csv", encoding="ISO-8859-1")#.dropna()
#print(ab.shape)
#ab.isnull().sum()
ab_noNaN = ab.drop(['Weight_oz','Width','Height'], axis=1).dropna()
ab_noNaN['Pub year'] = ab_noNaN['Pub year'].astype(int)
ab_noNaN['NumPages'] = ab_noNaN['NumPages'].astype(int)
ab_noNaN['Hard_or_Paper'] = ab_noNaN['Hard_or_Paper'].astype("category")
#print(ab_noNaN.shape)
#ab_noNaN.dtypes
ab_noNaN
````

````python
# train-test framework
import numpy as np
np.random.seed(131)

training_indecies = ab_noNaN.sample(frac=0.80, replace=False).index.sort_values()
testing_indecies = ab_noNaN.index[~ab_noNaN.index.isin(training_indecies)]
testing_indecies
````

````python
training_indecies
````

````python
# import statsmodels.api as sm
import statsmodels.formula.api as smf

simple_linear_regression_model = smf.ols('Q("Amazon Price") ~ Q("List Price")', 
                                         data=ab_noNaN)
model_fit = simple_linear_regression_model.fit() # creates line of best fit
model_fit.summary()
````

````python
SLRM_1 = smf.ols('Q("Amazon Price") ~ Q("List Price")', 
                                         data=ab_noNaN.loc[training_indecies,:])
SLRM_1_fit = SLRM_1.fit() # creates line of best fit
SLRM_1_fit.summary()
````

````python
np.corrcoef(SLRM_1_fit.predict(ab_noNaN.loc[testing_indecies,:]),
            ab_noNaN.loc[testing_indecies,"Amazon Price"])[0,1]**2
````

````python
SLRM_2 = smf.ols('Q("Amazon Price") ~ C(Hard_or_Paper)', 
                                         data=ab_noNaN.loc[training_indecies,:])
SLRM_2_fit = SLRM_2.fit() # cr.eates line of best fit
SLRM_2_fit.summary()
````

````python
SLRM_3 = smf.ols('Q("Amazon Price") ~ I(Q("List Price")**2)', 
                                         data=ab_noNaN.loc[training_indecies,:])
SLRM_3_fit = SLRM_3.fit() # creates line of best fit
SLRM_3_fit.summary()
````

````python
MLRM_4 = smf.ols('Q("Amazon Price") ~ Q("List Price") + I(Q("List Price")**2)', 
                                         data=ab_noNaN.loc[training_indecies,:])
MLRM_4_fit = MLRM_4.fit() # creates line of best fit
MLRM_4_fit.summary()
````

````python
MLRM_4_fit.model.formula
#MLRM_4_fit.model.endog
MLRM_4_fit.model.exog
````

````python
MLRM_5 = smf.ols('Q("Amazon Price") ~ Q("List Price")*C(Hard_or_Paper) + I(Q("List Price")**2)', 
                                         data=ab_noNaN.loc[training_indecies,:])
MLRM_5_fit = MLRM_5.fit() # creates line of best fit
MLRM_5_fit.summary()
````

````python
MLRM_6 = smf.ols('Q("Amazon Price") ~ Thick + NumPages + Q("List Price")*C(Hard_or_Paper) + I(Q("List Price")**2)', 
                                         data=ab_noNaN.loc[training_indecies,:])
MLRM_6_fit = MLRM_6.fit() # creates line of best fit
MLRM_6_fit.summary()
````

````python
MLRM_6_fit.model.formula
#MLRM_6_fit.model.exog.shape
MLRM_6_fit.model.exog_names
````

````python
MLRM_7 = smf.ols('Q("Amazon Price") ~ Thick + C(NumPages) + Q("List Price")*C(Hard_or_Paper) + I(Q("List Price")**2)', 
                                         data=ab_noNaN.loc[training_indecies,:])
MLRM_7_fit = MLRM_7.fit() # creates line of best fit
MLRM_7_fit.summary()
````

````python
ab_noNaN_training_data = ab_noNaN.loc[training_indecies,:].copy()
ab_noNaN_training_data['Publisher'] = ab_noNaN_training_data['Publisher'].astype('category').cat.set_categories(ab_noNaN['Publisher'].unique())
ab_noNaN_training_data['NumPages'] = ab_noNaN_training_data['NumPages'].astype('category').cat.set_categories(ab_noNaN['NumPages'].unique())
````

````python
MLRM_8 = smf.ols('Q("Amazon Price") ~ Thick + C(Publisher) + C(NumPages) + Q("List Price")*C(Hard_or_Paper) + I(Q("List Price")**2)', 
                                         data=ab_noNaN_training_data)#ab_noNaN.loc[training_indecies,:])
MLRM_8_fit = MLRM_8.fit() # creates line of best fit
MLRM_8_fit.summary()
````

````python
MLRM_8_fit.model.formula
MLRM_8_fit.model.exog.shape
MLRM_8_fit.model.exog_names
````

````python
# overfitting..,
np.corrcoef(MLRM_8_fit.predict(ab_noNaN.loc[testing_indecies,:]),
            ab_noNaN.loc[testing_indecies,"Amazon Price"])[0,1]**2
````

# Transition to `sklearn`

\begin{align*}
E[y] ={}& \beta_0 + \beta_1 \times \text{ListPrice} + \beta_2 \times 1_{[\text{'P'}]}(\text{Hard_or_Paper})\; + \\{}& \quad\;\;\; \beta_3 \times \text{ListPrice}\times 1_{[\text{'P'}]}(\text{Hard_or_Paper}) + \beta_4 \times \text{ListPrice}^2
\end{align*}
- not hardback: $\beta_0 + \beta_1 \times \text{ListPrice} + \beta_4 \times \text{ListPrice}^2$
- are hardback: $(\beta_0+\beta_2) + (\beta_1+\beta_3) \times \text{ListPrice} + \beta_4 \times \text{ListPrice}^2$

````python
from sklearn import tree
clf = tree.DecisionTreeClassifier()
clf
````

````python
X = ab_noNaN[['NumPages', 'Thick', 'List Price']]

clf = clf.fit(X.loc[training_indecies,:], y[training_indecies])
# Can't do... `formula = "Hard_or_Paper ~ Height + Width"`

_ = tree.plot_tree(clf)
````

````python
# _ = tree.plot_tree(clf)
import graphviz 
dot_data = tree.export_graphviz(clf, class_names=['Paperback', 'Hardback'], 
                                feature_names=X.columns,  
                                out_file=None, filled=True, rounded=True,
                                special_characters=True)  
graph = graphviz.Source(dot_data)  
graph 
````

````python
# A confusion matrix is a way to evaluate classification predictions
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm_disp = ConfusionMatrixDisplay(confusion_matrix(y[training_indecies].values, 
                                                  clf.predict(X.loc[training_indecies,:]), 
                                                  labels=[0,1]),
                                 display_labels=['P','H'])
_ = cm_disp.plot()
````

# Confusion Matrices are often presented in terms of FP/FN(TP/TN)

|         |Pred P(-) |Pred H(+) |
|---------|----------|----------|
|True P(-)|TrueNegative(TN)|FalsePositive(FP)|       
|True H(+)|FalseNegative(FN)|TruePositive(TP)|

- **Sensitivity** [proportion of correct predictions in the actually positive class]:<br>TP/(TP+FN), e.g., up above our sensitivity is ?/? = ?% sensitivity
- **Specificity** [proportion of correct predictions in the actually negative class]:<br>TN/(TN+FP), e.g., up above our specificity is ?/? = ?% specificity
- **Accuracy** [overall proportion of correct predictions]:<br>(TP+TN)/(TP+TN+FP+FN)[=N], e.g., up above our accurac is ?/? = ?%

Note that these should be "out of sample metrics" -- this should be performance on test data -- not training used to fit the model (which would over optimistically estimate these metric scores)

https://en.wikipedia.org/wiki/Confusion_matrix#Table_of_confusion

````python
cm_disp = ConfusionMatrixDisplay(confusion_matrix(y[testing_indecies].values, 
                                                  clf.predict(X.loc[testing_indecies,:]), 
                                                  labels=[0,1]),
                                 display_labels=['P','H'])
_ = cm_disp.plot()
````
